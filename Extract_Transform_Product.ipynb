```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# DATAFRAME EXAMPLE 1
```

```python
url = requests.get("https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets")
```

```python
url.status_code
```

```python
soup = BeautifulSoup(url.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
tag = soup.header.div.a.button.span
tag.string
```

```python
price = soup.find_all('h4', class_ = "price float-end card-title pull-right")
price_list=[]
for i in price:
    elem = i.text
    price_list.append(elem)
print(price_list)
    
```

```python
desc = soup.find_all("a", class_ = "title")
desc_list=[]
for i in desc:
    elem = i.text
    desc_list.append(elem)
print(desc_list)
```

```python
what = soup.find_all("p", class_ = "description card-text")
what_list=[]
for i in what:
    elem = i.text
    what_list.append(elem)
print(what_list)
```

```python
review = soup.find_all("p", class_ = "review-count float-end")
review_list=[]
for i in review:
    elem = i.text
    review_list.append(elem)
print(review_list)
```

```python
productdf = pd.DataFrame({"Name" : desc_list,
                          "Description" : what_list,
                          "Price" : price_list, 
                          "Reviews" : review_list})
```

```python
productdf
```

```python
productdf.to_csv("product.csv", index = False)
```
